Frequency Table for dice rolls

```python
import random
roll_freq = {}
num_rolls = 100000
for k in range(num_rolls):
  result = random.randint(1,6) + random.randint(1, 6)
  if result in roll_freq:
    roll_freq[result] += 1
  else:
    roll_freq[result] = 1

print(roll_freq)
for r in roll_freq:
  print(f'{r} occurred {roll_freq[r]} times.')

print()

# for r in sorted(roll_freq):
#   print(f'{r} occurred {roll_freq[r]} times.')

# print(sorted(roll_freq.keys()))

# print(roll_freq.keys())
# print(roll_freq.values())
# print(sorted(roll_freq))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
sides = 6
rolls =list(sorted(roll_freq.keys()))
pos = np.arange(2, 2 + len(rolls))

#popularity = [roll_freq[r] for r in roll_freq]
popularity = [roll_freq[r] for r in rolls]

plt.bar(rolls, popularity, align='center', color =['brown', 'orange', 'yellow', 'green', 'blue'] )
plt.xticks(pos, rolls)
plt.ylabel('Frequency')
plt.title(f'Frequency of rolls of a pair of {sides}-sided dice, {num_rolls} trials', alpha=0.8)

plt.show()
```

```python
import csv
import random

# This code will read the roster I exported from Sakai
roster = []
with open('M215RosterAll.csv', 'r') as csvfile:
    names_reader = csv.reader(csvfile, delimiter=',')
    for row in names_reader:
      roster.append(row)

# Examine the beginning of the roster and understand its structure
#print(roster[:5])
#print(len(roster))
##print(f'roster[0] is {roster[0]}')

# Filter from roster only the names that are in section 002
names002 = []

for entry in roster:
  if entry[1] == 'Section 002':
    names002.append(entry[0])

#print(names002)
# Extract only first names from the list and work with those
name = names002[8]
comma_loc = name.find(',')
first_name = name[comma_loc + 2:]
#print(first_name)

first_names = [name[name.find(',') + 2:] for name in names002]
#print(first_names)
# Set the group size
group_size = 4
grace_loc = first_names.index('Grace')
first_names[grace_loc] = 'Grace F'
# randomly select three people and put them in a group until there are no people left to assign
group_list = []
current_group = []

while len(first_names) > 0:
  # randomly choose a person and add then to group, removing them from list
  k = random.randrange(len(first_names))
  current_group.append(first_names.pop(k))
  if len(current_group) >= group_size:
    group_list.append(current_group)
    current_group = []

if len(current_group) > 0:
  group_list.append(current_group)
  current_group = []


#print(group_list)
#print(current_group)
# print the groups in a way that is easy to read
num = 1
for g in group_list:
  print(num, end = ':  ')
  for p in g:
    print(p, end = '  ')
  num += 1
  print()
  print('-----------------------')


# What happend to the original roster?
```
