Replace a for loop with a while loop

```python
running_sum = 0
upper_bound = 11
for k in range(upper_bound):
  running_sum += k**2
print(running_sum)

running_sum = 0
k = 0  # Initialize
while k < upper_bound:  # Continue while k is less than 10
  running_sum += k**2
  #k +=1  # update k
print(running_sum)
```

Prompt the user to enter an even integer greater than 4

```python
x = int(input('Enter an an even integer greater than 4'))
while x <= 4 or x % 2 ==1 : # x is a bad value
  # ask again
  if x <= 4:
    print(f'{x} is too small')
  if x % 2 == 1:
    print(f'{x} is not even')
  x = int(input('try again'))

print(f'{x} is good')
```

Prompt user to enter a value in a range, loop until user gives good input

```python
x = int(input('Enter an integer in the range 4 to 100, inclusive of endpoints'))

while x < 4 or x > 100:
  if x < 4:
    print(f'{x} is too small')
  if x > 100:
    print(f'{x} is too big')
  x = int(input('try again'))

print(f'4 <= {x} <= 100')
```

Roll a pair of fair dice UNTIL you get a 2; count how many rolls it takes

```python
import random
roll_count = 0
d1 = random.randint(1, 6)
d2 = random.randint(1, 6)
result = d1 + d2
roll_count += 1
target_result = 7
while result != target_result:
  #roll again
  d1 = random.randint(1, 6)
  d2 = random.randint(1, 6)
  result = d1 + d2
  roll_count += 1
print(f'It took {roll_count} rolls to get {result}')
```

Generate all of the fibonacci numbers less than 1000, without using a list

```python
f1 = 1
f2 = 1
print(f1, end = ' ')
print(f2, end = ' ')
f = f1 + f2
while f < 1000:
  print(f, end = ' ')
  f1 = f2
  f2 = f
  f = f1 + f2

print()
```
