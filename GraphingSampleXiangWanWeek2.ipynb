Code to display a graph and some points in different colors

```python
# import libraries and set up the figure

import matplotlib.pyplot as plt
import numpy as np  # numpy also has lots of built in math functions
import random

plt.figure(figsize = (6, 6))
import math
# plot a curve on the domain xmin <= x <= xmax
xmin = 0
xmax = 1.0
x = np.arange(xmin, xmax, 0.01)
plt.plot(x, np.sin(20*x), color = 'blue')

#plot some random points on the same figure
ymin = 0
ymax = 1
num_points = 12
for k in range(num_points):
    xc = random.uniform(xmin, xmax)
    yc = random.uniform(ymin, ymax)
    plt.plot(xc, yc, color = 'green', marker = 'o', markersize = 15)

# add a title to the plot
plt.title(f'A sine curve with {num_points} random points')

# add axes to the plot
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.show()


```
