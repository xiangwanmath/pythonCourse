```python
# Use cmath for complex numbers
import cmath

print("cmath.sqrt(-8) =", cmath.sqrt(-8))
print("cmath.sqrt(20) =", cmath.sqrt(20))
print("cmath.sqrt(8 + 5j) =", cmath.sqrt(8 + 5j))
```

```python
r = cmath.sqrt(-8)
print("cmath.sqrt(-8) =", r)

#using round on the real and imaginary parts to clean up the display
print("cmath.sqrt(-8) =", round(r.real, 3), '+', round(r.imag, 3), 'j', sep = '')
r2 = cmath.sqrt(8 + 5j)
print("cmath.sqrt(8 + 5j) =", round(r2.real, 3), '+', round(r2.imag, 3), 'j', sep = ' ')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure()

xmin = -5
xmax = 5
x_values = np.arange(xmin, xmax, 0.1)
y_values =[ math.sin(x) for x in x_values]
z_values = [math.cos(x) for x in x_values]
plt.plot(x_values, y_values, color = 'blue')
plt.plot(x_values, z_values, color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np  # numpy also has lots of built in math functions

plt.figure()

xmin = -5
xmax = 5
x = np.arange(xmin, xmax, 0.1)
plt.plot(x, np.sin(x), color = 'green')

plt.show()
```
