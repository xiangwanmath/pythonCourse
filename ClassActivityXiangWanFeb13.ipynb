1.  The G-concentration in your string (the number of G’s in the string)/(length of string)
2.  A substring of your string of length 20 starting at position 3 and containing every other character from the original string until you get to length 20
3.  The longest sequence of consecutive T’s in your string
4.  The substring of your string of length 10 that has the highest G-concentration
5.  The first occurrence of the pattern, GAC
6.  The last occurrence of the pattern, GAC



```python
import random
desired_length = 50
nucleotides = 'ACTG'
randomDNA_list = []
for k in range(desired_length):
  randomDNA_list.append(random.choice(nucleotides))
randomDNA_str = ''.join(randomDNA_list)
print(randomDNA_str)



```
