Working with lists and strings

Some slicing - you can slice a list or a string

```python
list1 = [2**k for k in range(11)]
print(f'list1 is {list1}')

list2 = list1[::-1]  # makes a reverse copy of the list
print(list2)

list3 = list1[3:8:2] # this is a distinct copy
print(f'list1[3:8:2] is {list3}')
list4 = list1
print(f'list4 is an alias for list 1 {list4}')
list3[0] = 'Z'
print(f'list1: {list1}')
print(f'list3: {list3}')
list5 = list1[:]
list4[0] = 'XYZ'
print(list1)
print(list4)
print(list5)
```

Extract user ids from a list of email addresses

```python
email_list = ['chaught@luc.edu', 'abc@google.com', 'ghoughton@luc.edu', 'lbabc@wiley.com', 'helpdesk@luc.edu' , 'no_at_symbol_here']

user_id_list = []

for email in email_list:
  at_loc = email.find('@')
  if at_loc >= 0:
    user_id = email[0:at_loc]
    user_id_list.append(user_id)

print(user_id_list)
```

More succinct code

```python
email_list = ['chaught@luc.edu', 'ghoughton@luc.edu', 'lbabc@wiley.com', 'helpdesk@luc.edu' , 'no_at_symbol_here']

user_id_list = []

for email in email_list:
  if '@' in email:
    user_id_list.append(email[0:email.find('@')])

print(user_id_list)
```

Count how many rolls it takes to get doubles

```python

import random

#initialize
r1 = random.randint(1, 6)
r2 = random.randint(1,6)
roll_count = 1

while r1 != r2:
  r1 = random.randint(1, 6)
  r2 = random.randint(1,6)
  roll_count += 1


print(f'It took {roll_count} rolls to get doubles')

```

Display the prime factorization of an integer

```python
x = 125
y = x
# initialize
f = 2
factor_list = []

while x > 1:
  #check if f goes into x
  if x % f == 0:
    factor_list.append(f)
    x = x // f
  else:
    f += 1


print(factor_list)
if len(factor_list)  == 1:
  factor_list.insert(0, 1)

str_factor_list = [str(f) for f in factor_list]
print(str_factor_list)
result = ' * '.join(str_factor_list)
print(f' {y}  = {result}')

```
