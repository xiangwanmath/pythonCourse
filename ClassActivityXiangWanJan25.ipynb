Generating Random Numbers

```python
import random
print("random.random()")
# Generate 5 random real numbers in the interval [0, 1)
for k in range(5):
  print(random.random())

print("\nrandom.uniform(5, 12)")
# Generate 5 random real numbers in the interval [5, 12)
for k in range(5):
  print(random.uniform(5, 12))

print("\nrandom.randint(1, 6)")
# Generate 5 random integers in the interval [1, 6] (1 and 6 are both included)
# Good for simulating the roll of a 6-sided die
for k in range(15):
  print(random.randint(1, 6), end = ' ')

print("\nrandom.randrange(1, 6)")
# Generate 5 random integers in the interval [1, 6) (6 is not possible)
for k in range(15):
  print(random.randrange(1, 6), end = ' ')

print("\nrandom.choice('ACTG')")
# Generate 5 random selections from the string "ACTG".  Good for generating random dna
for k in range(5):
  print(random.choice('ACTG'), end = ' ')
```

Test a trig identity at random points in the interval [0, 2pi)
Demonstrates difficulty in using testing for equality within float type

```python
import random, math
count = 0
numTrials = 100
for k in range(numTrials):
  x = random.uniform(0, 2*math.pi)
  if math.sin(x)**2 + math.cos(x)**2 != 1:
    print("identity fails at x =" , x, math.sin(x)**2 + math.cos(x)**2)
    count = count + 1

print(count, 'failures in', numTrials, 'trials')
```

Use round to test for equality with 8 digits

```python
import random, math
count = 0
numTrials = 1000
for k in range(numTrials):
  x = random.uniform(0, 2*math.pi)
  if round(math.sin(x)**2 + math.cos(x)**2, 10) != 1:
    print("identity fails at x =" , x, math.sin(x)**2 + math.cos(x)**2)
    count = count + 1

print(count, 'failures in', numTrials, 'trials')
```
