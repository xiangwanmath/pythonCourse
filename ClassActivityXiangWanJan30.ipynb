Solutions for the factoring drill problem
1.  generate and display a random number for c
2.  prompt and read user values for a, b
3.  give correct feedback
4.  enclose all this in a for loop
5.  add in scoring

```python
import random
score = 0
for k in range(5):
  c = random.randint(4, 100)
  print(f'Factor the number {c} into a * b')

  a = int(input('a = '))
  b = int(input('b = '))

  if c % a == 0:
    print(f'Good job, {a} is a factor of {c}')
    if 1 < a < c:  #a is a non trivial factor
      score += 1
  if a == b:
    if c//a % b == 0:
      print(f'Good job, {b} is a factor of {c//a}')
      if 1 < b < c:
        score += 1
  elif c % b == 0:
    print(f'Good job, {b} is a factor of {c}')
    if 1 < b < c:
      score += 1
  if a * b == c:
    print(f'Great! {a} * {b} = {c}')
    score += 2
  else:
    print(f'Sorry, {a} * {b} is not = {c}')

print(score)
```

Processing a string with a for loop

```python
capString = ''
for ch in myString:
    capString += ch.upper()# converts to upper case, concatenation
    print(capString)


print(myString)
print(capString)
print(myString.upper())
```

```python
numString = ''
digits = '0123456789'
for c in myString:
    if c in digits:
        numString += c
print(myString)
print(numString)

zString = numString.zfill(10)
print(zString)
zString2 = myString.zfill(100)
print(zString2)
```

Roll a pair of fair 6-sided dice n times and report the fraction of the rolls that we doubles.  Use this to approximate the probability of rolling doubles

```python
import random

num_trials = 10000
count = 0
for k in range(num_trials):
  # Roll two dice
  d1 = random.randint(1, 6)
  d2 = random.randint(1, 6)
  #What are we looking for?  Doubles?  Try something else
  if d1 == d2:
    count += 1

print(f'There were {count} rolls of doubles out of {num_trials} rolls.  \nEstimated probability: {count/num_trials:.3f}')
```
