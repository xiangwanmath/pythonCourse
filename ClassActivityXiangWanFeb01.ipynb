Use a for-loop to sum a series

```python
# Geometric series with r = 1/2 converges fast to 2
num_terms = 200
running_sum = 0
for n in range(num_terms):
  next_term = 1/2**n
  running_sum += next_term

print(f'The sum is {running_sum} using {num_terms} terms')
```

```python
# Alternating harmonic series, converges slowly to ln(2)
num_terms = 200000
running_sum = 0
for n in range(1, num_terms):
  next_term = (-1)**(n-1) / n
  running_sum += next_term

print(f'The sum is {running_sum} using {num_terms} terms')
```
