```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

majors =['Math', 'Data Sci', 'BioInform.',  'Econ', "Biology", "Other"]
pos = np.arange(len(majors))
popularity = [1, 1, 12, 1, 1, 1]

plt.bar(pos, popularity, align='center', color =['brown', 'orange', 'yellow', 'green', 'blue'] )
plt.xticks(pos, majors)
plt.ylabel('Popularity')
plt.title('Majors in our Math 215 class', alpha=0.8)
plt.tick_params( bottom = False, top = False, labelbottom = True, labeltop=False, labelleft=True, labelright=False)

plt.show()
```
