Dynamic typing and polymorphism

```python
# Demonstrate parameter passing
import math

def f(x):  # what is the type of x? x only exists inside f
    print('Inside function f, type(x) = ', type(x))  # x = True is boolean
    return 3*x  # x is promoted to be an integer, True = 1, False = 0

def main():
    print('f(5) = ', f(5))
    print('f(math.pi) = ', f(math.pi))
    print('f(\'abc\') = ', f('abc'))
    print('f([5, 6, 7]) = ', f([5, 6, 7]))
    print('f((5, 6, 7)) = ', f((5, 6, 7)))
    print('f(True) = ', f(True))
    print('f(False) = ', f(False))
    print('int(True) = ', int(True))
    print('int(False) = ', int(False))
    #print('x = ', x)


main()
```

Function calls with mutable types

```python
def replace_first_v1(aseq, newVal):  # input parameters : aseq, newVal
    # This works for lists, fails for strings
    # Notice there is no return value
    aseq[0]=newVal  # assignment is illegal for strings - they are immutable

def replace_first_v2(aseq, newVal):  #inpute parameters aseq, newVal
    aseq = newVal + aseq[1:]
    return aseq

def main():
    myList = [1,2, 3]
    print(f'Initially myList is : {myList}')
    replace_first_v1(myList, 'A')
    print(f' after the call to replace_first_v1 myList is : {myList}')
    print()

    myStr = 'xyz'
    print(f'Initially myStr is : {myStr}')
    answer = replace_first_v2(myStr, 'A')
    print(f' after the call to replace_first_v2 myStr is : {myStr}')
    print(f' after the call to replace_first_v2 answer is : {answer}')
    print()

    myList = [1,2, 3]
    print(f'Initially myList is : {myList}')
    ans = replace_first_v2(myList, ['A'])
    print(f' after the call to replace_first_v2 myList is : {myList}')
    print(f' after the call to replace_first_v2 ans is : {ans}')
    print()

main()
```

Functions as parameters to other functions

```python
import math
def sum_series(start, stop, a):
  # Add the terms a(start) + a(start + 1) + ... + a(stop)
  answer = 0
  for k in range(start, stop+1):
    answer += a(k)
  return answer

def f(n):
  return 1/math.factorial(n)

def g(n):
  return (-1)**n/(2*n+1)

print(sum_series(0, 20, f))
print(sum_series(0, 10, lambda x:1/math.factorial(x)))

print(4* sum_series(0, 1000, g))
```
