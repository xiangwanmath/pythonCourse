```python
# Scoring Password Strength

import string
# Some helpful string library constants
'''
print(string.ascii_lowercase)
print(string.ascii_uppercase)
print(string.digits)
print(string.punctuation)
'''

def num_char(pwd_str):
  # Input:  pwd_str:  a password string
  # Returns: an integer score = 4 * number of characters
  return 4*len(pwd_str)

def ucLetter(pwd_str):
  # Input:  pwd_str:  a password string
  # Returns: an integer score 2 * (length of pwd_str - number of upper case letters)

  #uc = sum([1 for sym in pwd_str if sym in string.ascii_uppercase])
  counter = 0
  for s in pwd_str:
    if s in string.ascii_uppercase:
      counter += 1
  if counter > 0:
    return (len(pwd_str) - counter)*2
  else:
    return 0

def numbers(pwd_str):#Alena
  #Input: a password string
  #Returns:  an integer score 4 * number of numbers
  n = 0
  for ch in pwd_str:
    if ch in string.digits:
      n+=1

  return 4 * n


def symbols(pwd_str):#Giana
  #Input: a password string
  #Returns:  an integer score 6 * number of symbols
  symbol_count = 0
  for ch in pwd_str:
    if ch in string.punctuation:
      symbol_count += 1
  return symbol_count * 6


def middle(pwd_str):#Paige
  #Input: a password string
  #Returns:  an integer score 2 * number of middle numbers or letters
  # Middle means not first or last num or letter
  not_middle = '09azAZ'
  n = 0
  for ch in pwd_str:
    if ch not in not_middle:
      n += 1
  return 2 * n

def letters_only(pwd_str):#Shawn
  #Input: a password string
  #Returns:  an integer score (-1) * length if letters only, 0 ow
  non_letters = 0
  for x in pwd_str:
    if x in string.digits or x in string.punctuation:
      non_letters += 1
      break
  if non_letters > 0:
    return 0
  else:
    return -1*len(pwd_str)

def consecutive_numbers(pwd_str):#Richa
  #Input: a password string
  #Returns:  -n, where the consecutive sequence of numbers has length n+1, 0 ow
  current_run = 0
  running_total = 0
  for ch in pwd_str:
    if ch in string.digits:
      current_run +=1
      if current_run >1:
        running_total +=1
    else:
      current_run = 0


  return (-2) * running_total


def sequential_numbers(pwd_str):#To do on Oct 17
  #Input: a password string
  #Returns:  - 3 n, where the sequential sequence of numbers has length n+2 >= 3, 0 ow
  # What about decreaasing sequences?
  # What about multiple sequences?



  return 0

def score(pwd):
  print(f'\n\nTo do:  add more features to the score\n\n')
  answer = 0
  answer += num_char(pwd)
  answer += ucLetter(pwd)
  answer += numbers(pwd)
  answer += symbols(pwd)
  answer += middle(pwd)
  answer += letters_only(pwd)
  answer += consecutive_numbers(pwd)
  answer += sequential_numbers(pwd)
  return answer

def interp(score):
  answer = ''
  if score > 80:
    answer = 'Excellent'
  elif score > 60:
    answer = 'Adequate security'
  elif score > 40:
    answer = 'Weak security'
  else:
    answer = "Very weak password"

  return answer

def main():
  pwdStr = input("Enter a password: ")#'abc'
  s = score(pwdStr)
  print(f'{pwdStr} has a score of {s}')
  print(f'Interpretation: {interp(s)}')

main()
```

```python
#Numerical Integration
import math
def left_riemann_sum(f, a, b, n):
  dx = (b-a)/n
  answer = 0
  x = a
  for k in range(n):
    answer += f(x)*dx
    x += dx
  return answer

def trapezoid_rule(f, a, b, n):
  dx = (b-a)/n
  answer = 0
  x = a
  for k in range(n):
    answer += (f(x) + f(x + dx)) / 2 * dx
    x += dx
  return answer

def g(x):
  return x**2

def h(x):
  return math.exp(-(x**2))

# Write another example not using lambda notation
print(f'left_riemann_sum(lambda x: x**2, 0, 10, 1000) = {left_riemann_sum(lambda x: x**2, 0, 10, 1000)}')
print(f'trapezoid_rule(lambda x: x**2, 0, 10, 1000)   = {trapezoid_rule(lambda x: x**2, 0, 10, 1000)}')
print(f'Calculus answer = 10**3/3                     = {10**3/3}')

print(f'left_riemann_sum(g, 0, 10, 1000) = {left_riemann_sum(g, 0, 5, 1000)}')
print(f'Calculus answer = 5**3/3                     = {5**3/3}')

print(f'left_riemann_sum(h, 0, 5, 10) = {left_riemann_sum(h, 0, 5, 10)}')
print(f'left_riemann_sum(h, 0, 5, 100) = {left_riemann_sum(h, 0, 5, 100)}')
print(f'left_riemann_sum(h, 0, 5, 10000) = {left_riemann_sum(h, 0, 5, 10000)}')
```
